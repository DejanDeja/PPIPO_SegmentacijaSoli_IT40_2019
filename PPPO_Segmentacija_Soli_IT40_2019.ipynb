**SEGMENTACIJA**

Segmentacija fotografija je proces deljenja slike na različite segmente ili regione radi identifikacije i analize specifičnih objekata ili regija unutar slike. Osnovni cilj segmentacije je prepoznati granice između različitih objekata ili regiona kako bismo omogućili razumevanje i tumačenje sadržaja slike na nivou pojedinačnih delova.

Svaka slika predstavlja kolekciju ili skup različitih piksela, a segmentacija omogućava grupisanje piksela koji imaju slične atribute. To nam omogućava da istražimo važne segmente slike i usredsredimo se na njih tokom obrade, umesto da istovremeno obradimo celu sliku što može dovesti do gubitka relevantnih informacija.

Proces segmentacije može se izvesti na više nivoa detaljnosti. Grubo segmentiranje može dovesti do formiranja većih regiona, dok fino-granularna segmentacija se može baviti pikselima na nivou pojedinačnih detalja.

Ukratko, segmentacija slike predstavlja moćan alat u oblasti digitalne obrade slika i računarske vizije. Podelom slike na segmente, možemo selektivno obrađivati važne delove slike i olakšati analizu i prepoznavanje objekata u slikama.

**VRSTE SEGMENTACIJE**

Segmentacija slika može se klasifikovati u nekoliko osnovnih vrsta, koje se koriste za razdvajanje objekata ili regiona unutar slike, pri čemu svaka metoda koristi određene kriterijume ili karakteristike za identifikaciju tih segmenata.

Vrste segmentacije obuhvataju:



1.   Segmentacija na osnovu boje: Ova metoda koristi informacije o boji piksela kako bi razlikovala različite objekte ili regione na slici. To se postiže korišćenjem metoda klasterovanja, kao što je k-sredina algoritam, kako bi se grupisali pikseli sličnih boja.
2.   Segmentacija na osnovu ivica: Ova metoda pronalazi prelaze intenziteta ili promene u slikovnom sadržaju kako bi identifikovala ivice objekata. Primer metode koja se koristi za segmentaciju na osnovu ivica je Kani algoritam.
3.   Segmentacija na osnovu teksture: Ova metoda analizira varijacije u teksturi piksela kako bi identifikovala različite regione. Koriste se tehnike kao što su Gaborovi filteri ili Haralikove karakteristike za izdvajanje teksturalnih osobina.
4.   Segmentacija na osnovu regiona: Ova metoda koristi informacije o konzistentnosti i homogenosti piksela unutar regiona. Algoritmi poput algoritma rasta regiona ili algoritma razdvajanja regiona spadaju u ovu kategoriju.
5.   Segmentacija na osnovu aktivnosti: Ova metoda se često koristi u video segmentaciji i koristi informacije o promenama u pokretu ili aktivnosti kako bi se razlikovali objekti ili regioni.
6.   Segmentacija na osnovu dubokog učenja: Ova metoda koristi duboke neuronske mreže za segmentaciju slika. Konvolutivne neuronske mreže (CNN) se često koriste za ovu svrhu, a popularne arhitekture uključuju U-Net, Mask R-CNN i SegNet.

U osnovnim vrstama segmentacije, moguće je kombinovati različite metode kako bi se postigli najbolji rezultati. Primena odgovarajuće vrste segmentacije zavisi od konkretnih zahteva i svojstava slika koje se obrađuju.

**NAJKORIŠĆENIJE TEHNIKE**

1.   **Segmentacija pragom (Thresholding)**: Ova tehnika je jednostavna i često se koristi kada su objekti i pozadina dobro odvojeni po intenzitetu. Prag se postavlja kako bi se odredila granica između objekata i pozadine. Pikseli sa vrednostima iznad praga klasifikuju se kao objekti, dok se pikseli ispod praga klasifikuju kao pozadina. Prag se može postaviti ručno ili automatski, na primer, korišćenjem OTSU metode za automatsko određivanje praga.
2.   **Segmentacija klasterovanjem (Clustering)**: Klasterovanje je tehnika grupisanja piksela na osnovu sličnosti u prostoru karakteristika. K-means algoritam je jedan od najpoznatijih algoritama klasterovanja koji se koristi za segmentaciju slika. Algoritam iterativno grupiše piksele u K klastera tako da se minimizuje suma kvadrata udaljenosti između piksela i centara klastera. Ova tehnika je posebno korisna kada želimo izvršiti segmentaciju slika na osnovu boje ili intenziteta.
3.   **Segmentacija preklapanjem regiona (Region merging/splitting)**: Ova tehnika se koristi za spajanje ili razdvajanje susednih regiona na slici na osnovu njihove sličnosti ili različitosti. Algoritam rasta regiona (region growing) počinje sa pojedinačnim pikselom ili malom grupom piksela i proširuje region na susedne piksele koji imaju slične karakteristike. Algoritam razdvajanja-regija (region splitting) deli početni region na manje regione na osnovu razlika u karakteristikama. Ova tehnika se često koristi za segmentaciju objekata koji su povezani ili se delimično preklapaju.
4.   **Segmentacija na osnovu ivica (Edge-based methods)**: Ova tehnika koristi informacije o intenzitetu piksela kako bi se identifikovale ivice objekata na slici. Canny algoritam je jedan od najpoznatijih algoritama za otkrivanje ivica. Ovaj algoritam identifikuje značajne promene intenziteta piksela i vrši postupak filtriranja i povezivanja ivica kako bi se dobio tačan prikaz ivica objekata na slici. Ove ivice se zatim mogu koristiti za segmentaciju.

**U-NET ARHITEKTURA KONVOLUCIJSKE NEURONSKE MREŽE**

U okviru projekta, koristili smo U-Net arhitekturu konvolucijske neuronske mreže za brŽu i preciznu semantičku segmentaciju slika. U-Net je posebno dizajniran za ovu svrhu i dobio je ime po svojoj karakterističnoj "slovo U" ili "slovo V" strukturi.

U arhitekturi U-Net-a, postoje dva ključna dela: "enkoder" i "dekoder". Enkoder se sastoji od serije konvolucijskih slojeva koji služe za izdvajanje karakteristika iz ulazne slike i postepeno smanjenje prostorne rezolucije. Ovi slojevi uključuju konvolucijske slojeve, slojeve sažimanja (*pooling*) ili slojeve sažimanja kao što je max pooling. Ova faza omogućava izdvajanje apstraktnijih karakteristika.

Nakon enkodera, sledi dekoder koji koristi konvolucijske slojeve i up-sampling slojeve kako bi povećao prostornu rezoluciju i rekonstruisao segmentiranu sliku. Ova faza omogućava rekonstrukciju detalja i postupno vraćanje na originalnu rezoluciju slike.

Posebno važne su "skip veze" (*skip connections*) koje povezuju kontraktni i ekspanzioni put. Ove veze omogućavaju prenos niskih i visokih nivoa karakteristika između različitih slojeva, što pomaže u kombinovanju lokalnih i globalnih informacija i poboljšava kvalitetu segmentacije.

U-Net arhitektura je postala popularna u oblasti segmentacije slika zbog svoje sposobnosti da nauči i rekonstruiše detaljne informacije iz slika, čak i kada su objekti mali ili slabo kontrastni. Ova arhitektura je takođe efikasna u radu s malim skupovima podataka jer koristi "augmentaciju podataka" koja generiše varijacije postojećih slika kako bi poboljšala sposobnost mreže da generalizuje.

Ovaj kod kreira 4x4 mrežu za prikazivanje slika iz trening skupa podataka. U svakom redu se prikazuju 3 slike: originalna slika, maska i preklapanje originalne slike sa maskom. Slike se nasumično biraju iz trening skupa, a pre nego što se prikažu, primenjuje se transformacija kako bi se slike prikazale u crno-belom formatu. Ovakav prikaz omogućava vizuelni pregled kako izgledaju slike i maske koje će biti korištene u obučavanju modela za segmentaciju slika.

```python
class Segmentation_Dataset(Dataset):
    def __init__(self, image_path, mask_path):
        self.image_path = image_path
        self.mask_path = mask_path

        images = glob.glob(image_path + '/*.png')

        samples = []

        for image in images:
            samples.append(image.split('/')[-1].split('.')[0])

        self.samples = samples

        self.transforms = T.Compose([T.Grayscale(), T.ToTensor()])

    def __getitem__(self, idx):
        image = Image.open(os.path.join(self.image_path, self.samples[idx] + '.png'))
        mask = Image.open(os.path.join(self.mask_path, self.samples[idx] + '.png'))

        return self.transforms(image), self.transforms(mask)

    def __len__(self):
        return len(self.samples)
```

Ovaj kod definiše prilagođeni dataset za segmentaciju slika, nazvan Segmentation_Dataset. Konstruktor klase inicijalizuje putanje do foldera sa slikama i maskama, kao i transformacije koje će biti primenjene na slike. Metoda getitem omogućava pristupanje paru slika i maski na osnovu indeksa, pri čemu se slike otvaraju, transformišu u crno-bele i pretvaraju u tenzore pre nego što se vrate. Metoda len vraća dužinu dataset-a, odnosno broj uzoraka u njemu. Ova klasa omogućava pravilno učitavanje i transformisanje slika i maski za dalje korišćenje u procesu obučavanja ili testiranja modela za segmentaciju.

```python
train_dataset = Segmentation_Dataset(IMAGE_TRAIN_PATH, MASK_TRAIN_PATH)
```

Ovde kreiramo instancu klase Segmentation_Dataset sa nazivom train_dataset. Prilikom kreiranja, prosleđuju se putanje do foldera sa trening slikama (IMAGE_TRAIN_PATH) i foldera sa maskama za trening slika (MASK_TRAIN_PATH). Nakon izvršavanja ovog koda, train_dataset će predstavljati prilagođeni dataset koji sadrži originalne slike i odgovarajuće maske za trening, spremne za dalje korišćenje u obučavanju modela za segmentaciju.

```python
class ConvolutionBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride=1, activation=True):
        super(ConvolutionBlock, self).__init__()

        self.activation = activation

        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride=stride, padding=kernel_size // 2)

        self.norm = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        x = self.conv(x)
        x = self.norm(x)

        if self.activation:
            return F.relu(x)
        else:
            return x
```

Ovaj kod definiše klasu ConvolutionBlock koja predstavlja osnovni konvoluciono-normacijski blok za neuronsku mrežu. Konstruktor klase inicijalizuje blok sa brojem ulaznih i izlaznih kanala, veličinom jezgra konvolucije, korakom (stride), i opcionalnom aktivacionom funkcijom. Metoda forward() izvršava prolazak kroz konvolucioni sloj, normalizaciju i, opciono, primenu aktivacione funkcije (ReLU). Ova klasa se može koristiti za kreiranje slojeva konvolucione neuronske mreže sa ili bez aktivacije, kao i sa normalizacijom.

```python
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3):
        super(ResidualBlock, self).__init__()

        self.norm = nn.BatchNorm2d(in_channels)

        self.conv1 = ConvolutionBlock(in_channels, out_channels, kernel_size)
        self.conv2 = ConvolutionBlock(out_channels, out_channels, kernel_size, activation=False)

    def forward(self, x):
        residual = x

        x = F.relu(x)

        x = self.norm(x)
        x = self.conv1(x)
        x = self.conv2(x)

        return x
```

Ovaj kod definiše klasu ResidualBlock koja predstavlja rezidualni blok (residual block) za neuronsku mrežu. Rezidualni blok je ključna komponenta u ResNet arhitekturi koja omogućava efikasno treniranje dubokih neuronskih mreža. Konstruktor klase inicijalizuje blok sa brojem ulaznih i izlaznih kanala, veličinom jezgra konvolucije (podrazumevano 3). Metoda forward() izvršava prolazak kroz residualni blok, gde se originalni ulaz (x) čuva kao rezidual (residual), a zatim se primenjuju dva konvoluciona bloka (ConvolutionBlock) sa slojevima konvolucije, normalizacije i opcionalnom aktivacionom funkcijom (ReLU). Na kraju, izlaz residualnog bloka se vraća.

```python
class DeconvolutionBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=2, stride=2):
        super(DeconvolutionBlock, self).__init__()

        self.deconv = nn.ConvTranspose2d(in_channels, out_channels, kernel_size=kernel_size, stride=stride)

    def forward(self, x1, x2):
        xd = self.deconv(x1)

        x = torch.cat([xd,x2], dim=1)

        return x
```

Ovaj kod definiše klasu DeconvolutionBlock koja predstavlja dekonvolucionu (transponovanu konvoluciju) blok za neuronsku mrežu, često korišćen u arhitekturama za segmentaciju slika. Konstruktor klase inicijalizuje blok sa brojem ulaznih i izlaznih kanala, veličinom jezgra dekonvolucije (podrazumevano 2) i veličinom koraka (stride) dekonvolucije (podrazumevano 2). Metoda forward() izvršava dekonvoluciju nad ulazom x1, a zatim spaja rezultat dekonvolucije sa drugim ulazom x2, koristeći konkatenaciju po drugoj dimenziji (dim=1). Na kraju, kombinovani izlaz se vraća.

```python
class UNet(nn.Module):
    def __init__(self, filters=16, dropout=0.5):
        super(UNet, self).__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(1, filters, 3, padding=1),
            ResidualBlock(filters, filters),
            ResidualBlock(filters, filters),
            nn.ReLU()
        )

        self.conv2 = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout / 2),
            nn.Conv2d(filters, filters * 2, 3, padding=1),
            ResidualBlock(filters * 2, filters * 2),
            ResidualBlock(filters * 2, filters * 2),
            nn.ReLU()
        )

        self.conv3 = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout),
            nn.Conv2d(filters * 2, filters * 4, 3, padding=1),
            ResidualBlock(filters * 4, filters * 4),
            ResidualBlock(filters * 4, filters * 4),
            nn.ReLU()
        )

        self.conv4 = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout),
            nn.Conv2d(filters * 4, filters * 8, 3, padding=1),
            ResidualBlock(filters * 8, filters * 8),
            ResidualBlock(filters * 8, filters * 8),
            nn.ReLU()
        )

        self.middle = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout),
            nn.Conv2d(filters * 8, filters * 16, 3, padding=3 // 2),
            ResidualBlock(filters * 16, filters * 16),
            ResidualBlock(filters * 16, filters * 16),
            nn.ReLU()
        )

        self.deconv4 = DeconvolutionBlock(filters * 16, filters * 8, 2)

        self.upconv4 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 16, filters * 8, 3, padding=1),
            ResidualBlock(filters * 8, filters * 8),
            ResidualBlock(filters * 8, filters * 8),
            nn.ReLU()
        )

        self.deconv3 = DeconvolutionBlock(filters * 8, filters * 4, 3)

        self.upconv3 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 8, filters * 4, 3, padding=1),
            ResidualBlock(filters * 4, filters * 4),
            ResidualBlock(filters * 4, filters * 4),
            nn.ReLU()
        )

        self.deconv2 = DeconvolutionBlock(filters * 4, filters * 2, 2)

        self.upconv2 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 4, filters * 2, 3, padding=1),
            ResidualBlock(filters * 2, filters * 2),
            ResidualBlock(filters * 2, filters * 2),
            nn.ReLU()
        )

        self.deconv1 = DeconvolutionBlock(filters * 2, filters, 3)

        self.upconv1 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 2, filters, 3, padding=1),
            ResidualBlock(filters, filters),
            ResidualBlock(filters, filters),
            nn.ReLU(),
            nn.Dropout(dropout / 2),
            nn.Conv2d(filters, 1, 3, padding=1)
        )

    def forward(self, x):
        conv1 = self.conv1(x)
        # 101 -> 50
        conv2 = self.conv2(conv1)
        # 50 -> 25
        conv3 = self.conv3(conv2)
        # 25 -> 12
        conv4 = self.conv4(conv3)
        # 12 - 6
        x = self.middle(conv4)
        # 6 -> 12
        x = self.deconv4(x, conv4)
        x = self.upconv4(x)
        # 12 -> 25
        x = self.deconv3(x, conv3)
        x = self.upconv3(x)
        # 25 -> 50
        x = self.deconv2(x, conv2)
        x = self.upconv2(x)
        # 50 -> 101
        x = self.deconv1(x, conv1)
        x = self.upconv1(x)

        return x
```

Ovaj kod definiše arhitekturu UNet neuronske mreže za segmentaciju slika, koja se sastoji od osnovnih konvolucionih i dekonvolucionih blokova, sa dodatkom rezidualnih veza za poboljšanje performansi mreže. Arhitektura ima ukupno 16 slojeva i koristi dropout kao regularizaciju kako bi se smanjio rizik od prenaučenosti.

```python
def iou_score(outputs, labels):
    A = labels.squeeze().bool()

    pred = torch.where(outputs < 0., torch.zeros_like(outputs), torch.ones_like(outputs))

    B = pred.squeeze().bool()

    intersection = (A & B).float().sum((1,2))

    union = (A | B).float().sum((1,2))

    iou = (intersection + 1e-6) / (union + 1e-6)

    return iou
```

Ova funkcija izračunava IoU (Intersection over Union) skor za segmentaciju izlaznih predikcija (outputs) i oznaka (labels) na temelju definicije IoU metrike.

```python
def train_one_batch(model, x, y):
    x, y = x.to(device), y.to(device)

    outputs = model(x)

    loss = loss_fn(outputs, y)
    iou = iou_score(outputs, y).mean()

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    return loss.item(), iou.item()
```

Ova funkcija trenira model (neuronsku mrežu) za jedan mini-batch podataka (ulaz x i odgovarajuće oznake y), izračunava gubitak (loss) i IoU skor (Intersection over Union) na temelju izlaznih predikcija modela, primenjuje povratno širenje (backpropagation) kako bi osvežila težine modela i vraća gubitak i IoU kao vrednosti koje se koriste za praćenje performansi treninga.

```python
def time_since(since):
    now = time.time()

    s = now - since
    m = math.floor(s / 60)
    s -= m * 60

    return '%dm %ds' % (m,s)
```

Ova funkcija izračunava vreme proteklo od trenutka "since" do trenutka izvršavanja funkcije, a zatim vraća vreme u formatu "mm:ss" (minuti:sekunde).

```python
NUM_EPOCHS = 100
BATCH_SIZE = 64

model = UNet().to(device)

model.train()

optimizer = torch.optim.Adam(model.parameters())
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max')

loss_fn = nn.BCEWithLogitsLoss()

train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

steps = train_dataset.__len__() // BATCH_SIZE

start = time.time()

train_loss = []

train_iou = []

for epoch in range(1, NUM_EPOCHS+1):
    print('Epoch {}/{}'.format(epoch, NUM_EPOCHS))

    running_iou = []
    running_loss = []

    for step, (x,y) in enumerate(train_dataloader):
        loss, iou = train_one_batch(model, x, y)

        running_iou.append(iou)
        running_loss.append(loss)

        print('\r{:6.1f} %\tloss {:8.4f}\tIoU {:8.4f}'.format(100 * (step + 1) / steps, loss, iou), end='')

    print('\r{:6.1f} %\tloss {:8.4f}\tIoU {:8.4f}\t{}'.format(100 * (step + 1) / steps, np.mean(running_loss), np.mean(running_iou), time_since(start)))

    scheduler.step(np.mean(running_iou))

    train_loss.append(loss)
    train_iou.append(iou)
```

Ovaj kod predstavlja petlju za treniranje UNet modela za segmentaciju slika. Model se trenira kroz 100 epoha, koristeći Adam optimizator i Binary Cross Entropy With Logits Loss funkciju za gubitak. Tokom svake epohe, model se obučava na podacima iz train_dataloader-a, a rezultati gubitka i IoU (Intersection over Union) za svaku iteraciju se beleže i ispisuju na ekranu. Takođe, koristi se ReduceLROnPlateau scheduler kako bi se dinamički prilagodila stopa učenja tokom obučavanja, na osnovu promene IoU vrednosti.

```python
plt.plot(train_loss, label='loss')
plt.plot(train_iou, label='IoU')

plt.xlabel('epoch')
plt.ylabel('metric')

plt.legend()

plt.show()
```

Ovaj kod koristi biblioteku matplotlib za prikazivanje grafika metrike gubitka (loss) i metrike Intersection over Union (IoU) tokom obučavanja UNet modela za segmentaciju slika. Na x-osi se nalaze epohe, dok se na y-osi nalaze vrednosti gubitka i IoU. Legend-a prikazuje koja kriva odgovara kojoj metrici, a plt.show() prikazuje grafik na ekranu.

```python
def rle_encode(im):
    pixels = im.flatten(order='F')
    pixels = np.concatenate([[0], pixels, [0]])

    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1

    runs[1::2] -= runs[::2]

    return ' '.join(str(x) for x in runs)

transforms = T.Compose([T.Grayscale(), T.ToTensor()])
```

Ovaj kod definiše funkciju rle_encode, koja se koristi za enkodiranje segmentirane maske u format run-length encoding (RLE) koji se često koristi za kompaktno predstavljanje maske u formatu koji se lako može deliti i čitati. Takođe, definiše se transformacija slika kroz kompoziciju transformacija transforms koja uključuje pretvaranje slike u crno-bele (grayscale) i konverziju u tenzor (torch.Tensor).

```python
sub_df = pd.read_csv(SAMPLE_SUBMISSION_PATH)

n = sub_df.shape[0]

rle_mask = []

for idx in range(n):
    sample = sub_df['id'][idx]

    image = Image.open(os.path.join(IMAGE_TEST_PATH + '/{}.png'.format(sample)))
    image = transforms(image)

    out = model(image.unsqueeze(0)).squeeze()

    pred = torch.where(out < 0., torch.zeros_like(out), torch.ones_like(out))

    rle_mask.append(rle_encode(pred.numpy()))

sub_df['rle_mask'] = rle_mask
```
